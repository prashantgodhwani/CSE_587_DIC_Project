## Handling incorrect latitude and longitude values

Since our customer base resides in the US, we remove data with incorrect latitude and longitude values

```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

geometry = [Point(xy) for xy in zip(business['longitude'], business['latitude'])]
gdf = GeoDataFrame(business, geometry=geometry)   

#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(10, 6)), marker='o', color='red', markersize=20);
```

```python
def is_in_us(lat, long):
    if lat < 24.5271 or lat > 49.7662:
        return False
    if long < -124.7844 or long > -66.9514:
        return False
    return True

# filter rows where coordinates are in US
business = business[business.apply(lambda row: is_in_us(row['latitude'], row['longitude']), axis=1)]
business
```

```python
import langid
chunksize = 100000

# create an empty list to store the chunks of data
data = []

# read the data in chunks and process each chunk
for chunk in pd.read_csv("yelp_review.csv", chunksize=chunksize):
    # detect the language of the reviews in the current chunk using langid
    lang_list = [langid.classify(text)[0] for text in chunk['text']]

    # create a boolean mask that checks for non-English reviews in the current chunk
    mask = pd.Series(lang_list).apply(lambda x: x != 'en')

    # remove non-English reviews from the current chunk
    chunk = chunk[~mask]

    # append the processed chunk to the list of data
    data.append(chunk)

# concatenate the list of processed data into a single DataFrame
df_reviews = pd.concat(data)

# display the updated DataFrame
print(df_reviews)
```
